```python
ds=df.dropna(axis=0)
df.isnull().sum()
```

```python
x=df.drop(['Width'],axis=1)
x
```

```python
y=df['Width']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size =0.33,random_state = 0)
```

```python
print("xtrain shape : ", xtrain.shape)
print("xtest shape : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape : ", ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain,ytrain)
y_pred = regressor.predict(xtest)
```

```python
train_score=regressor.score(xtrain,ytrain)
test_score=regressor.score(xtest,ytest) 
```

```python
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest,y_pred)
print("Mean Square Error : ", mse)
print("Mean Absolute Error : ", mae)
```

```python
train_accuracy=regressor.score(xtrain,ytrain)
print("train accuracy=",train_accuracy)
```

```python
test_accuracy=regressor.score(xtest,ytest)
print("test accuracy=",test_accuracy)
```
